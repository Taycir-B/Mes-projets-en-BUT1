```python
# Nécessaire pour éviter les problèmes de session
class Connexion(object):
    def __init__(self, login, password):
        self.conn = oracledb.connect(
            user=login,
            password=password,
            host="oracle.iut-orsay.fr",
            port=1521,
            sid="etudom",
        )
        self.conn.autocommit = False

    def __enter__(self):
        self.conn.autocommit = False
        return self.conn

    def __exit__(self, *args):
        self.conn.close()
```

```python
import pandas as pd
from IPython.display import display

query_total_athletes = f"""
    SELECT COUNT(DISTINCT pa.IdAthlete) as Total_Athletes
FROM (
    SELECT pi.IdAthlete
    FROM participation_individuelle pi
    INNER JOIN evenement ev ON pi.IdEvent = ev.IdEvenement
    INNER JOIN hote h ON ev.IdHote = h.IdHote
    WHERE h.LibelleHote = '1992 Winter Olympics'
    
    UNION
    
    SELECT ce.IdAthlete
    FROM composition_equipe ce
    INNER JOIN participation_equipe pe ON ce.IdEquipe = pe.IdEquipe
    INNER JOIN evenement ev ON pe.IdEvenement = ev.IdEvenement
    INNER JOIN hote h ON ev.IdHote = h.IdHote
    WHERE h.LibelleHote = '1992 Winter Olympics'
) pa
"""
data_total_athletes = requete_vers_dataframe(conn, query_total_athletes)

TOTAL_ATHLETES = data_total_athletes['TOTAL_ATHLETES'].iloc[0]
print(f"Nombre d’athlètes participant à ces JO：{TOTAL_ATHLETES}")

query_athletes_by_gender = f"""
   SELECT a.Genre, COUNT(DISTINCT a.IdAthlete) as Total_Athletes
FROM ATHLETE a
INNER JOIN (
    SELECT pi.IdAthlete
    FROM participation_individuelle pi
    INNER JOIN evenement ev ON pi.IdEvent = ev.IdEvenement
    INNER JOIN hote h ON ev.IdHote = h.IdHote
    WHERE h.LibelleHote = '1992 Winter Olympics'
    
    UNION
    
    SELECT ce.IdAthlete
    FROM composition_equipe ce
    INNER JOIN participation_equipe pe ON ce.IdEquipe = pe.IdEquipe
    INNER JOIN evenement ev ON pe.IdEvenement = ev.IdEvenement
    INNER JOIN hote h ON ev.IdHote = h.IdHote
    WHERE h.LibelleHote = '1992 Winter Olympics'
) pa ON a.IdAthlete = pa.IdAthlete
GROUP BY a.Genre

"""
data_athletes_by_gender = requete_vers_dataframe(conn, query_athletes_by_gender)
print("Répartition par genre：")
display(data_athletes_by_gender)

query_athletes_by_country = f"""
    SELECT n.CodeNOC, COUNT(DISTINCT a.IdAthlete) as Total_Athletes
    FROM ATHLETE a
    INNER JOIN (
        SELECT pi.IdAthlete, pi.NOC
        FROM participation_individuelle pi
        INNER JOIN evenement ev ON pi.IdEvent = ev.IdEvenement
        INNER JOIN hote h ON ev.IdHote = h.IdHote
        WHERE h.LibelleHote = '1992 Winter Olympics'
        
        UNION
        
        SELECT ce.IdAthlete, e.NOC
        FROM composition_equipe ce
        INNER JOIN participation_equipe pe ON ce.IdEquipe = pe.IdEquipe
        INNER JOIN equipe e ON ce.IdEquipe = e.IdEquipe
        INNER JOIN evenement ev ON pe.IdEvenement = ev.IdEvenement
        INNER JOIN hote h ON ev.IdHote = h.IdHote
        WHERE h.LibelleHote = '1992 Winter Olympics'
    ) pa ON a.IdAthlete = pa.IdAthlete
    INNER JOIN noc n ON pa.NOC = n.CodeNOC
    GROUP BY n.CodeNOC
    ORDER BY Total_Athletes DESC
"""
data_athletes_by_country = requete_vers_dataframe(conn, query_athletes_by_country)
print("Par pays：")
display(data_athletes_by_country)
```

```python
query = """
SELECT
    a.NomAthlete,
    a.PrenomAthlete,
    COUNT(p.Medaille) AS Nombre_Total,
    SUM(CASE WHEN p.Medaille = 'Gold' THEN 1 ELSE 0 END) AS Nombre_Or,
    SUM(CASE WHEN p.Medaille = 'Silver' THEN 1 ELSE 0 END) AS Nombre_Argent,
    SUM(CASE WHEN p.Medaille = 'Bronze' THEN 1 ELSE 0 END) AS Nombre_Bronze
FROM PARTICIPATION_INDIVIDUELLE p
JOIN ATHLETE a ON p.IdAthlete = a.IdAthlete
JOIN EVENEMENT e ON p.IdEvent = e.IdEvenement
JOIN HOTE h ON e.IdHote = h.IdHote
WHERE h.LibelleHote = '1992 Winter Olympics'
GROUP BY a.NomAthlete, a.PrenomAthlete
ORDER BY Nombre_Total DESC, Nombre_Or DESC, Nombre_Argent DESC, Nombre_Bronze DESC
FETCH FIRST 10 ROWS ONLY
"""

Medaille_TOP10_INDIVIDUEL = requete_vers_dataframe(conn, query)

Medaille_TOP10_INDIVIDUEL
```

```python
query2 = """
SELECT
    n.NomNOC AS "NOC",
    SUM(CASE WHEN p.Medaille = 'Gold' THEN 1 ELSE 0 END) AS "Or",
    SUM(CASE WHEN p.Medaille = 'Silver' THEN 1 ELSE 0 END) AS "Argent",
    SUM(CASE WHEN p.Medaille = 'Bronze' THEN 1 ELSE 0 END) AS "Bronze",
    COUNT(p.Medaille) AS "Total_Medailles"
FROM
    PARTICIPATION_INDIVIDUELLE p
JOIN
    ATHLETE a ON p.IdAthlete = a.IdAthlete
JOIN
    EVENEMENT e ON p.IdEvent = e.IdEvenement
JOIN
    HOTE h ON e.IdHote = h.IdHote
JOIN
    NOC n ON p.NOC = n.codeNOC
WHERE
    h.LibelleHote = '1992 Winter Olympics' AND p.Medaille IS NOT NULL
GROUP BY
    n.NomNOC
ORDER BY
    "Total_Medailles" DESC, "Or" DESC, "Argent" DESC, "Bronze" DESC
FETCH FIRST 10 ROWS ONLY
"""


Medaille_NOC_TOP10_INDIVIDUEL = requete_vers_dataframe(conn, query2)
Medaille_NOC_TOP10_INDIVIDUEL
```

```python
# On donne ici l'exemple de comment ajouter un événement. On a stocké dans id_hote l'identifiant des JO sur lesquels on travaille
conn.begin()  # Démarre une nouvelle transaction
# On ouvre un curseur, qui est une sorte de pointeur vers une table de résultats
curseur = conn.cursor()
# On passe la requête paramétrée et les paramètres (s'il y en a) au curseur
curseur.execute(f"SELECT MAX(IDEVENEMENT) FROM {SCHEMA}.EVENEMENT")
# S'il n'y a pas d'erreur, on peut demander à récupérer les résultats : 
# - curseur.fetchone() récupère la ligne sur laquelle pointe le curseur, renvoie le résultat sous forme de tuple, et avance à la ligne suivante (si elle existe)
# - curseur.fetchall() boucle sur toutes les lignes restantes et renvoie une liste de tuples, ou une liste vide s'il n'y a plus de résultats
# - curseur.fetchmany(nb) renvoie les résultats des nb lignes suivantes sous forme de liste de tuples
idev = curseur.fetchone()[0]  # Renvoie un tuple, on veut le premier élément du tuple
idev += 1
curseur.execute(f"INSERT INTO {SCHEMA}.EVENEMENT VALUES(:idev, :nomev, :statutev, :codedisc, :idhote)", {'idev': idev, 'nomev': 'Individual, Men', 'statutev': 'Olympic', 'codedisc': 'SPS', 'idhote': id_hote})
conn.rollback()  # On annule, ce qui est au dessus est pour l'exemple, vous devrez utiliser commit() pour valider la transaction.
```

### Résultats des médailles

Pour terminer, saisir les résultats des médailles. En 1992, l'épreuve n'était pas une discipline officielle, il n'y avait donc pas de médaille, vous utiliserez donc le podium final comme base pour déterminer les médailles d'or, argent et de bronze.

```python
# Faire ici les insertions de données pour les résultats dans la table PARTICIPATION
```

```python
query_max_idevenement = f"SELECT MAX(IDEVENEMENT) FROM {SCHEMA}.EVENEMENT"
with conn.cursor() as curseur:
    curseur.execute(query_max_idevenement)
    idev = curseur.fetchone()[0]
    idev += 1

query_insert_event = f"""
    INSERT INTO {SCHEMA}.EVENEMENT (IDEVENEMENT, NOMEVENEMENT, STATUTEVENEMENT, CODEDISCIPLINE, IDHOTE)
    VALUES (:idev, :nomev, :statutev, :codedisc, :idhote)
"""
params_event = {
    "idev": idev,
    "nomev": 'Individual, Men',
    "statutev": 'Olympic',
    "codedisc": 'SPS',
    "idhote": id_hote
}

try:
    with conn.cursor() as curseur:
        curseur.execute(query_insert_event, params_event)

        athletes_results = [
            {"idathlete": 1, "medaille": "Gold", "idevent": idev, "resultat": 1},  # 假设 idathlete 为 1 的运动员获得金牌，排名 1
            {"idathlete": 2, "medaille": "Silver", "idevent": idev, "resultat": 2},  # 假设 idathlete 为 2 的运动员获得银牌，排名 2
            {"idathlete": 3, "medaille": "Bronze", "idevent": idev, "resultat": 3},  # 假设 idathlete 为 3 的运动员获得铜牌，排名 3
        ]
        for result in athletes_results:
            query_insert_medals = f"""
                INSERT INTO {SCHEMA}.PARTICIPATION_INDIVIDUELLE (IDATHLETE, IDEVENT, MEDAILLE, RESULTAT)
                VALUES (:idathlete, :idevent, :medaille, :resultat)
            """
            curseur.execute(query_insert_medals, {
                "idathlete": result["idathlete"],
                "idevent": result["idevent"],
                "medaille": result["medaille"],
                "resultat": result["resultat"]
            })

        conn.commit()
        print("Les résultats des médailles ont été insérés avec succès.")

except Exception as e:
    conn.rollback()
    print(f"Une erreur s'est produite : {e}")
```
